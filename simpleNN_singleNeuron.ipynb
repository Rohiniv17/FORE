```python
# 1.0 Call libraries
import pandas as pd
import numpy as np
from  pathlib import Path

# 1.0.1
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# 2.1 Where is our data file:
pathToFolder = "/gdrive/MyDrive/red_wine.csv"
```

```python
# 2.2 Full file path:
path = Path(pathToFolder)
```

```python
# 3.0 Read data
red_wine = pd.read_csv(path)
red_wine.head()
```

```python
# 3.1
red_wine.shape   # (1599, 12)
red_wine.dtypes
```

```python
# 3.2 Get target and predictors:
y = red_wine.pop("quality")
X = red_wine
```

```python
# 4.0 Build model:

model = tf.keras.Sequential()
```

```python

# 4.1
model.add(layers.Input(11,) )
```

```python
model.add(layers.Dense(11,)  )
```

```python
model.add(layers.Dense(11, activation = "relu"))
```

```python
model.add(layers.Dense(770,) )
```

```python
model.add(layers.Dense(770, activation="tanh"))
```

```python
# 4.2
model.add(layers.Dense(1))
```

image.png

```python
# 4.2.1
weights, bias = model.weights
```

```python
# 4.2.2
weights
```

```python
# 4.2.3
bias
```

```python
#4.3 Compile
model.compile( loss = "mae")
```

```python
# 4.4 Train
model.fit(X,y, epochs = 1400 )
```

```python
x# 5.0 Predict
pred = model.predict(X)
```

```python
# 5.1 Round up decimal places
pred = np.round(pred)
```

```python
# 5.1.1
y.shape
```

```python
# 5.2 Compare:
np.sum(pred == y.values.reshape(1599,1))/1599
```
