```python
pathToFolder = "/gdrive/MyDrive/mobile_price.csv"
```

```python
path = Path(pathToFolder)
```

```python
mobile_price = pd.read_csv(path)
mobile_price.head()
```

```python
mobile_price.tail()
```

```python
mobile_price.shape
mobile_price.dtypes
```

```python
y = mobile_price.pop("ram")
X = mobile_price
```

```python
model = tf.keras.Sequential()
```

```python
model.add(layers.Input(20,) ) #This should be modified depending on the No. of coumns in X
```

```python
model.add(layers.Dense(12,)  )
```

```python
model.add(layers.Dense(12, activation = "relu"))
```

```python
model.add(layers.Dense(770,) )
```

```python
model.add(layers.Dense(770, activation="relu"))
```

```python
model.add(layers.Dense(1))
```

```python
model.compile( loss = "mae")
```

```python
model.fit(X,y, epochs = 100 )
```

```python
pred = model.predict(X)
```

```python
pred = np.round(pred)
```
